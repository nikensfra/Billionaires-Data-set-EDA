11. ***Melakukan imputasi***. Hal ini dilakukan untuk data yang masih terdapat missing value baik data kategorik maupun kontinu. Untuk kategorik menggunakan nilai modus sebagai imputasinya sedangkan kontinu menggunakan nilai mean sebagai imputasinya.

### 4. Menganalisis perubahan demografi (usia dan jenis kelamin)  dan kekayaan miliarder selama bertahun-tahun.

```python
def remove_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))]
```

```python
# Buat kolom baru untuk kelompok usia
billionaires7['age_group'] = pd.cut(billionaires7['age'], bins=[0, 20, 30, 40, 50, 60, 70, 80, 90, 100], labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100'])
```

```python
billionaires7['finalWorth_no_outliers'] = billionaires7.groupby(['age_group','gender'])['finalWorth'].transform(remove_outliers)
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='age_group', y='finalWorth_no_outliers', hue='gender', data=billionaires7, palette='Set3')
plt.title('Distribusi Kekayaan Miliarder Berdasarkan Usia dan Gender')
plt.xlabel('Kelompok Usia')
plt.ylabel('Kekayaan (Final Worth)')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.swarmplot(x='age_group', y='finalWorth_no_outliers', hue='gender', data=billionaires7, palette='Set3')
plt.title('Distribusi Kekayaan Miliarder Berdasarkan Usia dan Gender')
plt.xlabel('Kelompok Usia')
plt.ylabel('Kekayaan (Final Worth)')
plt.legend(title='Gender')
plt.show()
```

**Kesimpulan**
Kekayaan Meningkat dengan Usia: Terdapat peningkatan jumlah kekayaan pada kelompok usia yang lebih tua, yaitu pada usia 80 & 90. Hal ini mungkin menunjukkan bahwa akumulasi kekayaan terjadi sepanjang karir seseorang, dengan puncak kekayaan biasanya tercapai pada usia yang lebih lanjut.

Perbedaan Gender: Dapat dilihat dari warna titik-titik yang mewakili gender (biru untuk M dan kuning untuk F), bahwa terdapat lebih banyak miliarder laki-laki dibandingkan perempuan.   
                                                                                                   
Kekayaan Ekstrim: Terdapat beberapa outlier, terutama di kelompok usia yang lebih tua, yang menunjukkan kekayaan yang sangat tinggi dibandingkan dengan yang lain. Ini bisa menandakan adanya beberapa individu yang memiliki kekayaan yang sangat besar, yang mungkin merupakan pendiri perusahaan besar atau mereka yang mewarisi kekayaan.

Kekayaan di Usia Muda: Terdapat beberapa individu di kelompok usia muda (21-30) yang sudah memiliki kekayaan yang signifikan.

Penurunan pada Usia Sangat Lanjut: Pada kelompok usia 91-100, tampaknya terjadi penurunan jumlah miliarder dan tingkat kekayaan mereka. Hal ini bisa diakibatkan oleh transisi kekayaan ke generasi berikutnya atau kemungkinan penurunan bisnis.
